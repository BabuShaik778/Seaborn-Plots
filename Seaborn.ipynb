```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading Csv
df = pd.read_csv('employee_data.csv')
df.info()
```

```python
# Hist

plt.hist(df['salary'])
plt.xlabel('Salary')
plt.ylabel('Salary distribution depend on ')
plt.title('Salary insights')
plt.show()
```

histogram graph for salary
*   The most eployees gets a salary of 60,0000
*   Very few people are getting a salary above 80,000
*   The number of employess getting salary from 20,000 to 60,000 increases and decreases above salary of 60,000 to 1,00,000. This means the employees who are getting salary of 20,000 to 60,000 is more and employee getting a salary above 60,000 to 1,00,000 is less.

```python
#Boxplot

sns.boxplot(data = df, x='department', y = 'salary', orient = 'v')
```


* Salary Range: HR has the most spread-out salaries, while Marketing and IT have more consistent pay.


 * Median Salary: HR has the highest average middle salary, while other departments have lower median salaries.


```python
# Barplot

sns.barplot(data = df, x='gender', y = 'satisfaction_score')
plt.title('averege satisfaction_score')
plt.xlabel('gender')
plt.ylabel('satisfaction_score')
plt.grid(True)
plt.show()
```

* Females rate satisfaction slightly higher than males
* Both genders score above 5 out of 6
* Satisfaction scores are very close, showing minimal gender difference

```python
# Countplot

sns.countplot(data = df, x = 'department')
plt.ylabel("No. of employees")
plt.show()
```

* Finance has the most employees (around 80)
* HR has the second-highest number of employees (around 72)
* Marketing and IT have fewer employees, with around 68-70 staff each

```python
# Scatterplot

sns.scatterplot(data = df, x = 'age' , y = 'salary', hue = 'gender')
plt.title("Relationship b/w Age & Salary")
plt.show()
```

* Salaries mostly range between 20,000 and 80,000
* Both males and females are represented across different ages
* No clear linear relationship between age and salary is visible

```python
# Heatmap

corr = df.corr(numeric_only = True)
sns.heatmap(corr, annot = True, cmap = 'rocket')
plt.title('Correlation Matrix')
plt.show()
```

  *  Most variables have very weak correlations (close to zero)
  * Age and salary show a tiny positive relationship - 0.049
  * No strong connections between age, salary, experience, and satisfaction  

```python
# Pairplot

sns.pairplot(df, hue='gender')
plt.suptitle("Pairplot of Numeric Variables", y=1.02)
plt.show()
```

```python
# Violinplot

sns.violinplot(data=df, x='left_company', y='satisfaction_score',color = 'hotpink')
plt.title("Total satisfaction score by employee left company")
plt.show()
```

* The plot shows satisfaction scores for employees who left and didn't leave the company
* The shapes are similar for both Yes (left company) and  No (stayed) suggesting consistent satisfaction levels
* The satisfaction scores appear to be distributed mostly between 4 and 8 groups

```python
# Implot

sns.lmplot(data=df, x='salary', y='years_experience', hue='left_company')
plt.title("Linear Relationship: salary vs Year")
plt.show()
```

* The salary spreading are 0 to 100000
* Here blue dots represents the who is stayed in comapnay
* Here orange dots represents the who is left in comapany
*


```python
# Swarmplot

sns.swarmplot(data=df, x='department', y='satisfaction_score')
plt.title("Satisfaction score Spread by Departments")
plt.show()
```

* IT has high satisfaction: Most scores are between 7 and 10.


* Finance has mixed satisfaction: Scores range from 2 to 10.


* HR is mostly satisfied: Scores are mainly between 6 and 10.

```python
# Stripplot

sns.stripplot(data=df, x='salary', y='years_experience', hue='left_company', jitter=True, dodge=True)
plt.title("Strip Plot: Salary vs Years of Experience")
plt.xlabel("Salary")
plt.ylabel("Years of Experience")
plt.legend(title="Left Company")
plt.show()
```

1. High salary doesn't stop people from leaving.
Even well-paid employees are quitting.


2. Mid-experience employees leave the most.
People with 4–7 years of experience are more likely to leave.


3. Freshers mostly stay.
Employees with 0–3 years experience usually don't leave.




```python
# Clustermap

df_num = df.select_dtypes(include='number')
sns.clustermap(df_num.corr(), annot=True, cmap="coolwarm")
plt.title("Employee Data - Correlation Clustermap", pad=70)
plt.show()
```

1. Salary and satisfaction have a very small positive link.


2. Experience doesn’t affect salary, age, or satisfaction much.


3. Age and satisfaction go up slightly together.

```python
# Regplot
sns.regplot(data=df, x='years_experience', y='salary', color='green')


plt.title("Regression Line: Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

1. Salary increases with experience - More years generally means higher pay
2. Huge salary differences - People with same experience earn very different amounts
3. Most people have 2-8 years experience - Few very new or very experienced workers

```python
# Lineplot

sns.lineplot(data=df, x='years_experience', y='salary')
plt.title("Line Plot: Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```

1. Line Plot:Salary jumps around wildly - Goes up and down randomly between $20k-$90k
2. Wrong chart type - Line plots shouldn't be used for this data
3. Makes no sense - No one's salary actually changes this erratically

```python
# Facetgrid

g = sns.FacetGrid(df, col="left_company", hue="gender")
g.map(sns.lineplot, "years_experience", "salary")
g.add_legend()

plt.subplots_adjust(top=0.8)
g.fig.suptitle("Line Plot: Salary vs Experience by Left Company & Gender")
plt.show()
```

1. Gender pay gap exists - Orange lines (male) generally run higher than blue lines (female)
2. Company leaving doesn't help much - Both panels show similar chaotic salary patterns
3. Line plots still misleading - Same problem as before - connects unrelated data points creating false impressions of individual career paths

```python
# Kdeplot

sns.kdeplot(data=df, x='salary', fill=True, color='skyblue')
plt.title("KDE Plot: Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Density")
plt.show()
```

1. Most people earn $50,000-$60,000 - The peak shows salary concentration around this range
2. Normal salary spread - Bell-shaped curve from about $20,000 to $100,000
3. Few extreme earners - Very low density at both very low and very high salaries

```python
#Displot

sns.displot(data=df, x='salary', kde=True, color='pink')
plt.title("Displot: Salary Distribution with KDE")
plt.xlabel("Salary")
plt.ylabel("Count / Density")
plt.show()
```

1. Salary forms a bell-shaped curve (normal distribution).


2. 45,000 to 75,000 is the common range.


3. Very few extreme values (outliers).


```python
# Rugplot

sns.kdeplot(data=df, x='salary', fill=True, color='purple')
sns.rugplot(data=df, x='salary', color='black')
plt.title("KDE + Rug Plot: Salary")
plt.xlabel("Salary")
plt.ylabel("Density")
plt.show()
```

1. Most salaries are around 60,000.


2. Salary data is evenly spread (no skew).


3. Many salaries fall between 40,000 to 80,000.

```python
# Relplot

sns.relplot(data=df, x="years_experience", y="salary", kind="scatter")
plt.title("relplot: Salary vs Years of Experience")
plt.show()
```

1. No clear pattern – Salary doesn’t steadily increase with experience.


2. Wide spread – People with the same experience have very different salaries.


3. Some outliers – A few people earn much more or much less than others.

```python
# CatPlot

sns.catplot(data=df, x='department', y='salary', kind='box', height=5, aspect=1.5)
plt.title("Box Plot: Salary by Department")
plt.xticks(rotation=45)
plt.show()
```

1. Similar median salary across all departments.


2. IT department has the widest salary range.


3. Marketing and HR have more stable salaries

```python
# JointGrid

g = sns.JointGrid(data=df, x="years_experience", y="salary", space=0)
g.plot_joint(sns.scatterplot, color="purple")
g.plot_marginals(sns.histplot, color="skyblue", kde=True)
plt.suptitle("JointGrid: Salary vs Years of Experience", y=1.05)
plt.show()
```

 1. No strong link between experience and salary — the points are scattered randomly.


2. Most people have 4–8 years of experience and earn around 60,000.


3. Some outliers exist — few low-experience people earn high salaries, and some experienced earn less.
